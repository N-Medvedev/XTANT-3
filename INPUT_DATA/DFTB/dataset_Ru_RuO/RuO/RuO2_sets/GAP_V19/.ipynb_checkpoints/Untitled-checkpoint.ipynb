```python
import ase, ase.io
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
def rms_dict(x_ref, x_pred):
    """ Takes two datasets of the same shape and returns a dictionary containing RMS error data"""
    x_ref = np.array(x_ref)
    x_pred = np.array(x_pred)
    if np.shape(x_pred) != np.shape(x_ref):
        raise ValueError('WARNING: not matching shapes in rms')
    error_2 = (x_ref - x_pred) ** 2
    average = np.sqrt(np.average(error_2))
    std_ = np.sqrt(np.var(error_2))
    return {'rmse': average, 'std': std_}
```

```python
train = ase.io.read("quip_train.xyz", ":")
```

```python
ru_atom_dft = train[0].info['dft_energy']
ru_atom_gap = train[0].info['energy']
o_atom_dft = train[1].info['dft_energy']
o_atom_gap = train[1].info['energy']
```

```python
# test
ru_atom_dft = train[0].info['dft_energy']
ru_atom_gap = ru_atom_dft
o_atom_dft = train[1].info['dft_energy']
o_atom_gap = o_atom_dft
```

```python
dft = {"bulk":[],"001":[],"010":[],"011":[],"110":[],"111":[]}
gap = {"bulk":[],"001":[],"010":[],"011":[],"110":[],"111":[]}
for atom in train:
    if atom.info['structure'] == 'bulk':
        n_O = int(re.split('(\d+)', atom.get_chemical_formula())[1])
        n_Ru = int(re.split('(\d+)', atom.get_chemical_formula())[3])
        dft["bulk"].append( (atom.info['dft_energy'] - n_O * o_atom_dft - n_Ru * ru_atom_dft) / atom.get_number_of_atoms() )
        gap["bulk"].append( (atom.info['energy'] - n_O * o_atom_gap - n_Ru * ru_atom_gap) / atom.get_number_of_atoms() )
    elif atom.info['structure_info'].startswith('001'):
        n_O = int(re.split('(\d+)', atom.get_chemical_formula())[1])
        n_Ru = int(re.split('(\d+)', atom.get_chemical_formula())[3])
        dft["001"].append( (atom.info['dft_energy'] - n_O * o_atom_dft - n_Ru * ru_atom_dft) / atom.get_number_of_atoms() )
        gap["001"].append( (atom.info['energy'] - n_O * o_atom_gap - n_Ru * ru_atom_gap) / atom.get_number_of_atoms() )
    elif atom.info['structure_info'].startswith('010'):
        n_O = int(re.split('(\d+)', atom.get_chemical_formula())[1])
        n_Ru = int(re.split('(\d+)', atom.get_chemical_formula())[3])
        dft["010"].append( (atom.info['dft_energy'] - n_O * o_atom_dft - n_Ru * ru_atom_dft) / atom.get_number_of_atoms() )
        gap["010"].append( (atom.info['energy'] - n_O * o_atom_gap - n_Ru * ru_atom_gap) / atom.get_number_of_atoms() )
    elif atom.info['structure_info'].startswith('011'):
        n_O = int(re.split('(\d+)', atom.get_chemical_formula())[1])
        n_Ru = int(re.split('(\d+)', atom.get_chemical_formula())[3])
        dft["011"].append( (atom.info['dft_energy'] - n_O * o_atom_dft - n_Ru * ru_atom_dft) / atom.get_number_of_atoms() )
        gap["011"].append( (atom.info['energy'] - n_O * o_atom_gap - n_Ru * ru_atom_gap) / atom.get_number_of_atoms() )
    elif atom.info['structure_info'].startswith('110'):
        n_O = int(re.split('(\d+)', atom.get_chemical_formula())[1])
        n_Ru = int(re.split('(\d+)', atom.get_chemical_formula())[3])
        dft["110"].append( (atom.info['dft_energy'] - n_O * o_atom_dft - n_Ru * ru_atom_dft) / atom.get_number_of_atoms() )
        gap["110"].append( (atom.info['energy'] - n_O * o_atom_gap - n_Ru * ru_atom_gap) / atom.get_number_of_atoms() )
    elif atom.info['structure_info'].startswith('111'):
        n_O = int(re.split('(\d+)', atom.get_chemical_formula())[1])
        n_Ru = int(re.split('(\d+)', atom.get_chemical_formula())[3])
        dft["111"].append( (atom.info['dft_energy'] - n_O * o_atom_dft - n_Ru * ru_atom_dft) / atom.get_number_of_atoms() )
        gap["111"].append( (atom.info['energy'] - n_O * o_atom_gap - n_Ru * ru_atom_gap) / atom.get_number_of_atoms() )
    else:
        print(atom.info['structure_info'])
```

```python
plt.rcParams["figure.figsize"] = (6,5)
plt.rcParams["font.size"] = 13
plt.rcParams["font.family"] = "Arial"
plt.style.use('ggplot')
plt.rcParams["axes.grid"] = False
plt.rcParams["axes.facecolor"] = "white"
plt.rcParams["axes.edgecolor"] = "black"

ax = plt.subplot(1,1,1)

#colors = list("rgbcmyk")

types = dft.keys()
#zorders=[2,6,3,1,5,4]
zorders=[5,1,4,6,2,3]

for n, t in enumerate(types):
    ax.scatter(dft[t], gap[t], zorder=zorders[n], label=t)

elim=[-5, -1]
ax.plot(elim, elim, c='k', zorder=0, label='scale')
plt.xlim(-4.6, -3.5)
plt.ylim(-4.6, -3.5)
plt.xlabel("$E_{\mathrm{DFT}}^{\mathrm{coh}}$ [eV/atom]", fontsize=14)
plt.ylabel("$E_{\mathrm{GAP}}^{\mathrm{coh}}$ [eV/atom]", fontsize=14)
_rms = rms_dict(sum(dft.values(),[]), sum(gap.values(), []))
rmse_text = 'RMSE: ' + str(np.round(_rms['rmse'], 3)) + '$\pm$' + str(np.round(_rms['std'], 4)) + ' eV/atom'
ax.text(0.9, 0.1, rmse_text, transform=ax.transAxes, fontsize='large', horizontalalignment='right', verticalalignment='bottom')
handles, labels = ax.get_legend_handles_labels()
labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: t[0]))
ax.legend(handles, labels, prop={"size":12})
#plt.figure(figsize=(8,8), dpi=300)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.savefig("energy.png")
```

```python
dft_o = {"bulk":[],"001":[],"010":[],"011":[],"110":[],"111":[]}
gap_o = {"bulk":[],"001":[],"010":[],"011":[],"110":[],"111":[]}
dft_ru = {"bulk":[],"001":[],"010":[],"011":[],"110":[],"111":[]}
gap_ru = {"bulk":[],"001":[],"010":[],"011":[],"110":[],"111":[]}

#ax = plt.subplot(1,1,1)
fig, ax_list = plt.subplots(nrows=2, ncols=1, gridspec_kw={'hspace': 0.3})
fig.set_size_inches(6, 11.5)
ax_list = ax_list.flat[:]

for atom in train:
    if atom.info['structure'] == 'bulk':
        for n, a in enumerate(atom):
            if a.symbol == 'O':
                dft_o["bulk"].append( atom.arrays['dft_forces'][n] )
                gap_o["bulk"].append( atom.arrays['force'][n] )
            elif a.symbol == 'Ru':
                dft_ru["bulk"].append( atom.arrays['dft_forces'][n] )
                gap_ru["bulk"].append( atom.arrays['force'][n] )
    elif atom.info['structure_info'].startswith('001'):
        for n, a in enumerate(atom):
            if a.symbol == 'O':
                dft_o["001"].append( atom.arrays['dft_forces'][n] )
                gap_o["001"].append( atom.arrays['force'][n] )
            elif a.symbol == 'Ru':
                dft_ru["001"].append( atom.arrays['dft_forces'][n] )
                gap_ru["001"].append( atom.arrays['force'][n] )
    elif atom.info['structure_info'].startswith('010'):
        for n, a in enumerate(atom):
            if a.symbol == 'O':
                dft_o["010"].append( atom.arrays['dft_forces'][n] )
                gap_o["010"].append( atom.arrays['force'][n] )
            elif a.symbol == 'Ru':
                dft_ru["010"].append( atom.arrays['dft_forces'][n] )
                gap_ru["010"].append( atom.arrays['force'][n] )
    elif atom.info['structure_info'].startswith('011'):
        for n, a in enumerate(atom):
            if a.symbol == 'O':
                dft_o["011"].append( atom.arrays['dft_forces'][n] )
                gap_o["011"].append( atom.arrays['force'][n] )
            elif a.symbol == 'Ru':
                dft_ru["011"].append( atom.arrays['dft_forces'][n] )
                gap_ru["011"].append( atom.arrays['force'][n] )
    elif atom.info['structure_info'].startswith('110'):
        for n, a in enumerate(atom):
            if a.symbol == 'O':
                dft_o["110"].append( atom.arrays['dft_forces'][n] )
                gap_o["110"].append( atom.arrays['force'][n] )
            elif a.symbol == 'Ru':
                dft_ru["110"].append( atom.arrays['dft_forces'][n] )
                gap_ru["110"].append( atom.arrays['force'][n] )
    elif atom.info['structure_info'].startswith('111'):
        for n, a in enumerate(atom):
            if a.symbol == 'O':
                dft_o["111"].append( atom.arrays['dft_forces'][n] )
                gap_o["111"].append( atom.arrays['force'][n] )
            elif a.symbol == 'Ru':
                dft_ru["111"].append( atom.arrays['dft_forces'][n] )
                gap_ru["111"].append( atom.arrays['force'][n] )
    else:
        print(atom.info['structure_info'])

print(dft_o.keys())
#print(dft_o)
zorders=[5,1,4,6,2,3]
for n, key in enumerate(dft_o.keys()):
    ax_list[0].scatter(dft_o[key], gap_o[key], zorder=zorders[n], label=key)

for n, key in enumerate(dft_ru.keys()):
    ax_list[1].scatter(dft_ru[key], gap_ru[key], zorder=zorders[n], label=key)
    
elim=[-30, 30]
ax_list[0].plot(elim, elim, c='k', zorder=0, label='scale')
ax_list[1].plot(elim, elim, c='k', zorder=0, label='scale')
for a in range(2):
    handles, labels = ax_list[a].get_legend_handles_labels()
    labels, handles = zip(*sorted(zip(labels, handles), key=lambda t: t[0]))
    ax_list[a].legend(handles, labels, prop={"size":12})
    ax_list[a].set_xlim(-28,28)
    ax_list[a].set_ylim(-28,28)
    ax_list[a].set_xlabel("$F_{\mathrm{DFT}}$ [eV/$\mathrm{\AA}$]", fontsize=14)
    ax_list[a].set_ylabel("$F_{\mathrm{GAP}}$ [eV/$\mathrm{\AA}$]", fontsize=14)
    ax_list[a].tick_params(axis='both', which='major', labelsize=14)
    ax_list[a].tick_params(axis='both', which='major', labelsize=14)
    

_rms_o = rms_dict(sum(dft_o.values(),[]), sum(gap_o.values(), []))
rmse_text_o = 'RMSE: ' + str(np.round(_rms_o['rmse'], 3)) + '$\pm$' + str(np.round(_rms_o['std'], 4)) + ' eV/$\mathrm{\AA}$'
ax_list[0].text(0.9, 0.1, rmse_text_o, transform=ax_list[0].transAxes, fontsize='large', horizontalalignment='right', verticalalignment='bottom')

_rms_ru = rms_dict(sum(dft_ru.values(),[]), sum(gap_ru.values(), []))
rmse_text_ru = 'RMSE: ' + str(np.round(_rms_ru['rmse'], 3)) + '$\pm$' + str(np.round(_rms_ru['std'], 4)) + ' eV/$\mathrm{\AA}$'
ax_list[1].text(0.9, 0.1, rmse_text_ru, transform=ax_list[1].transAxes, fontsize='large', horizontalalignment='right', verticalalignment='bottom')

plt.savefig("force.png")
```
